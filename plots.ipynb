```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
token_abalations = pd.read_csv('/Users/vashisth/Documents/GitHub/Project-Visual-Prompt-Tuning/wandb_export_2024-04-26T20_40_15.915-04_00.csv', index_col=False)
token_abalations
```

```python
x = token_abalations['tokens']
y = token_abalations['test_acc']
print(f'x: {x}')
print(f'y: {y}')
```

```python
plt.figure(figsize=(8, 6))  # Adjust the figure size

plt.plot(x, y, color='#1f77b4', linewidth=2)
plt.scatter(x, y, color='#ff7f0e', marker='o', s=80)

plt.xlabel('Prompt Size', fontsize=14)
plt.ylabel('Test Accuracy', fontsize=14)
plt.title('Prompt Size vs Test Accuracy', fontsize=16, fontweight='bold')

plt.grid(True, linestyle='--', alpha=0.7)

plt.xticks(x, fontsize=12)
plt.yticks(fontsize=12)

plt.legend(fontsize=12, loc='lower right')

plt.tight_layout()  # Adjust the layout to prevent overlapping

plt.savefig('tokens_vs_test_acc.png', dpi=300)  # Increase the resolution
plt.show()
```
